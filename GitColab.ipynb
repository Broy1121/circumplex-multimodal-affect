````python

import pickle

with open("data/processed_mosei.pkl", "rb") as f:
    mosei = pickle.load(f)

sample_key = list(mosei.keys())[0]
sample_val = mosei[sample_key]

print("Top-level type:", type(mosei))
print("Sample key:", sample_key)
print("Value type:", type(sample_val))
print("Value length:", len(sample_val))

# Peek inside the list
for i, item in enumerate(sample_val):
    print(f"  [{i}] type={type(item).__name__}", end="")
    if hasattr(item, 'shape'):
        print(f"  shape={item.shape}", end="")
    elif isinstance(item, dict):
        print(f"  keys={list(item.keys())}", end="")
    elif isinstance(item, (list, tuple)):
        print(f"  len={len(item)}", end="")
    else:
        print(f"  val={item}", end="")
    print()


##What you're likely seeing** is that MOSEI is stored as a list of segments per video ID, where each segment is either a tuple or a sub-list like:
#
````

````python
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from torch.optim.lr_scheduler import CosineAnnealingLR
import warnings
warnings.filterwarnings("ignore", message="Mean of empty slice")

# ── Device ─────────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

# ═══════════════════════════════════════════════════════════
# 1. HELPERS
# ═══════════════════════════════════════════════════════════
def extract_va(label_features):
    valence = float(label_features[0])
    arousal = float(np.max(label_features[[1, 3, 4, 6]]))  # happiness, anger, surprise, fear
    return valence, arousal

def clean_features(features):
    features = features.copy()
    features[~np.isfinite(features)] = 0.0
    return features

def pad_sequence_2d(features, max_len):
    T, D = features.shape
    if T >= max_len:
        return features[:max_len]
    pad = np.zeros((max_len - T, D), dtype=np.float32)
    return np.vstack([features, pad])

MAX_TEXT   = 40
MAX_AUDIO  = 600
MAX_VISION = 200

# ═══════════════════════════════════════════════════════════
# 2. DATASET
# ═══════════════════════════════════════════════════════════
class MOSEITemporalDataset(Dataset):
    def __init__(self, mosei_dict):
        self.samples = []
        for video_id, utterances in mosei_dict.items():
            for seg in utterances:
                try:
                    labels_raw  = np.array(seg['Labels']).item()
                    covarep_raw = np.array(seg['COVAREP']).item()
                    facet_raw   = np.array(seg['FACET']).item()
                    wordvec_raw = np.array(seg['WORDVEC']).item()

                    text   = pad_sequence_2d(clean_features(wordvec_raw['features'].astype(np.float32)), MAX_TEXT)
                    audio  = pad_sequence_2d(clean_features(covarep_raw['features'].astype(np.float32)), MAX_AUDIO)
                    vision = pad_sequence_2d(clean_features(facet_raw['features'].astype(np.float32)),   MAX_VISION)

                    valence, arousal = extract_va(labels_raw['features'])
                    if not (np.isfinite(valence) and np.isfinite(arousal)):
                        continue

                    self.samples.append({
                        'text':    torch.tensor(text),
                        'audio':   torch.tensor(audio),
                        'vision':  torch.tensor(vision),
                        'valence': torch.tensor(valence, dtype=torch.float32),
                        'arousal': torch.tensor(arousal, dtype=torch.float32),
                    })
                except Exception as e:
                    print(f"Skipped {video_id}: {e}")

    def __len__(self): return len(self.samples)
    def __getitem__(self, idx): return self.samples[idx]

def collate_fn(batch):
    return {
        'text':    torch.stack([s['text']    for s in batch]),
        'audio':   torch.stack([s['audio']   for s in batch]),
        'vision':  torch.stack([s['vision']  for s in batch]),
        'valence': torch.stack([s['valence'] for s in batch]),
        'arousal': torch.stack([s['arousal'] for s in batch]),
    }

# Build
dataset = MOSEITemporalDataset(mosei)
print(f"Total utterances: {len(dataset)}")

s = dataset[0]
print(f"text={s['text'].shape} audio={s['audio'].shape} vision={s['vision'].shape}")
print(f"valence={s['valence'].item():.4f}  arousal={s['arousal'].item():.4f}")

# Splits
total   = len(dataset)
n_train = int(0.8 * total)
n_val   = int(0.1 * total)
n_test  = total - n_train - n_val

train_set, val_set, test_set = random_split(
    dataset, [n_train, n_val, n_test],
    generator=torch.Generator().manual_seed(42)
)

train_loader = DataLoader(train_set, batch_size=32, shuffle=True,  collate_fn=collate_fn)
val_loader   = DataLoader(val_set,   batch_size=32, shuffle=False, collate_fn=collate_fn)
test_loader  = DataLoader(test_set,  batch_size=32, shuffle=False, collate_fn=collate_fn)

print(f"Train={len(train_set)}  Val={len(val_set)}  Test={len(test_set)}")

# ═══════════════════════════════════════════════════════════
# 3. MODEL
# ═══════════════════════════════════════════════════════════
class TemporalEncoder(nn.Module):
    def __init__(self, input_dim, hidden_dim=128, num_layers=2, dropout=0.3):
        super().__init__()
        self.gru = nn.GRU(
            input_dim, hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.attn = nn.Linear(hidden_dim * 2, 1)
        self.proj = nn.Linear(hidden_dim * 2, hidden_dim)
        self.norm = nn.LayerNorm(hidden_dim)

    def forward(self, x):
        out, _  = self.gru(x)                              # (B, T, H*2)
        weights = torch.softmax(self.attn(out), dim=1)     # (B, T, 1)
        pooled  = (out * weights).sum(dim=1)               # (B, H*2)
        return self.norm(self.proj(pooled))                 # (B, H)

class MultimodalTemporalModel(nn.Module):
    def __init__(self, hidden_dim=128, n_heads=4, dropout=0.3):
        super().__init__()
        self.text_enc   = TemporalEncoder(300, hidden_dim, dropout=dropout)
        self.audio_enc  = TemporalEncoder(74,  hidden_dim, dropout=dropout)
        self.vision_enc = TemporalEncoder(35,  hidden_dim, dropout=dropout)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim, nhead=n_heads,
            dim_feedforward=256, dropout=dropout,
            batch_first=True
        )
        self.fusion  = nn.TransformerEncoder(encoder_layer, num_layers=2)
        self.va_head = nn.Sequential(
            nn.Linear(hidden_dim * 3, 64),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(64, 2)
        )

    def forward(self, text, audio, vision):
        t = self.text_enc(text).unsqueeze(1)
        a = self.audio_enc(audio).unsqueeze(1)
        v = self.vision_enc(vision).unsqueeze(1)
        tokens = torch.cat([t, a, v], dim=1)       # (B, 3, H)
        fused  = self.fusion(tokens)
        flat   = fused.reshape(fused.size(0), -1)
        out    = self.va_head(flat)
        return out[:, 0], out[:, 1]

# ═══════════════════════════════════════════════════════════
# 4. TRAINING UTILITIES
# ═══════════════════════════════════════════════════════════
def concordance_correlation_coefficient(pred, target):
    pred_mean, target_mean = pred.mean(), target.mean()
    covariance = ((pred - pred_mean) * (target - target_mean)).mean()
    ccc = (2 * covariance) / (
        pred.var() + target.var() + (pred_mean - target_mean) ** 2 + 1e-8
    )
    return ccc

def va_loss(v_pred, a_pred, v_true, a_true, alpha=0.5):
    mse_v = nn.functional.mse_loss(v_pred, v_true)
    mse_a = nn.functional.mse_loss(a_pred, a_true)
    ccc_v = 1 - concordance_correlation_coefficient(v_pred, v_true)
    ccc_a = 1 - concordance_correlation_coefficient(a_pred, a_true)
    return alpha * (mse_v + mse_a) + (1 - alpha) * (ccc_v + ccc_a)

def run_epoch(loader, train=True):
    model.train() if train else model.eval()
    total_loss, ccc_v_sum, ccc_a_sum = 0, 0, 0
    ctx = torch.enable_grad() if train else torch.no_grad()
    with ctx:
        for batch in loader:
            text   = batch['text'].to(device)
            audio  = batch['audio'].to(device)
            vision = batch['vision'].to(device)
            v_true = batch['valence'].to(device)
            a_true = batch['arousal'].to(device)

            v_pred, a_pred = model(text, audio, vision)
            loss = va_loss(v_pred, a_pred, v_true, a_true)

            if train:
                optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()

            total_loss += loss.item()
            ccc_v_sum  += concordance_correlation_coefficient(v_pred.detach(), v_true).item()
            ccc_a_sum  += concordance_correlation_coefficient(a_pred.detach(), a_true).item()

    n = len(loader)
    return total_loss / n, ccc_v_sum / n, ccc_a_sum / n

# ═══════════════════════════════════════════════════════════
# 5. SANITY CHECK — forward pass before training
# ═══════════════════════════════════════════════════════════
model     = MultimodalTemporalModel(hidden_dim=128).to(device)
optimizer = optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=30, eta_min=1e-6)

batch = next(iter(train_loader))
with torch.no_grad():
    vp, ap = model(batch['text'].to(device), batch['audio'].to(device), batch['vision'].to(device))

print(f"\nForward pass OK — v_pred:{vp.shape}  a_pred:{ap.shape}")
print(f"Params: {sum(p.numel() for p in model.parameters()):,}")
print(f"NaNs in output: {torch.isnan(vp).any() or torch.isnan(ap).any()}")
````

**flow in diagram:**
```
mosei.pkl
    ↓
MOSEITemporalDataset    (clean, pad, extract labels)
    ↓
random_split            (80/10/10)
    ↓
DataLoader              (batches of 32)
    ↓
MultimodalTemporalModel (GRU encoders + Transformer fusion)
    ↓
va_loss                 (MSE + CCC)
    ↓
AdamW optimizer         (updates weights)
    ↓
CosineAnnealingLR       (reduces learning rate over time)
    ↓
Training loop × 30      (saves best checkpoint)

````python
EPOCHS = 30
best_val_loss = float('inf')
history = []

for epoch in range(1, EPOCHS + 1):
    tr_loss, tr_ccc_v, tr_ccc_a = run_epoch(train_loader, train=True)
    vl_loss, vl_ccc_v, vl_ccc_a = run_epoch(val_loader,   train=False)
    scheduler.step()

    history.append({
        'epoch': epoch,
        'tr_loss': tr_loss, 'vl_loss': vl_loss,
        'tr_ccc_v': tr_ccc_v, 'vl_ccc_v': vl_ccc_v,
        'tr_ccc_a': tr_ccc_a, 'vl_ccc_a': vl_ccc_a,
    })

    if vl_loss < best_val_loss:
        best_val_loss = vl_loss
        torch.save(model.state_dict(), "best_temporal_model.pt")
        tag = " ← best"
    else:
        tag = ""

    if epoch % 5 == 0 or epoch == 1:
        print(
            f"Ep {epoch:02d} | "
            f"Loss {tr_loss:.4f}/{vl_loss:.4f} | "
            f"CCC-V {tr_ccc_v:.3f}/{vl_ccc_v:.3f} | "
            f"CCC-A {tr_ccc_a:.3f}/{vl_ccc_a:.3f}{tag}"
        )

print(f"\nBest val loss: {best_val_loss:.4f}")
````

````python
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader

def extract_va(label_features):
    """
    label_features: np.array shape (7,)
    Returns valence (float), arousal (float)
    """
    valence = float(label_features[0])  # sentiment score, range ~[-3, 3]
    # Arousal proxy: max activation among high-energy emotions
    arousal = float(np.max(label_features[[1, 3, 4, 6]]))  # happiness, anger, surprise, fear
    return valence, arousal

def pool_features(features, method='mean'):
    """Collapse (T, D) → (D,) via mean or max pooling."""
    if method == 'mean':
        return features.mean(axis=0)
    elif method == 'max':
        return features.max(axis=0)

class MOSEIDataset(Dataset):
    def __init__(self, mosei_dict, pool='mean'):
        self.samples = []

        for video_id, utterances in mosei_dict.items():
            for seg in utterances:
                try:
                    # Unwrap numpy object scalars
                    labels_raw = np.array(seg['Labels']).item()
                    covarep_raw = np.array(seg['COVAREP']).item()
                    facet_raw  = np.array(seg['FACET']).item()
                    wordvec_raw = np.array(seg['WORDVEC']).item()

                    # Extract features (T, D) and pool → (D,)
                    text   = pool_features(wordvec_raw['features'].astype(np.float32), pool)  # (300,)
                    audio  = pool_features(covarep_raw['features'].astype(np.float32), pool)  # (74,)
                    vision = pool_features(facet_raw['features'].astype(np.float32),   pool)  # (35,)

                    # Extract VA labels
                    label_feats = labels_raw['features']  # (7,)
                    valence, arousal = extract_va(label_feats)

                    self.samples.append({
                        'text':    torch.tensor(text),
                        'audio':   torch.tensor(audio),
                        'vision':  torch.tensor(vision),
                        'valence': torch.tensor(valence, dtype=torch.float32),
                        'arousal': torch.tensor(arousal, dtype=torch.float32),
                        'video_id': video_id,
                    })

                except Exception as e:
                    print(f"Skipped {video_id}: {e}")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        return self.samples[idx]


# ── Sanity check ──────────────────────────────────────────
dataset = MOSEIDataset(mosei)
print(f"Total utterances: {len(dataset)}")

s = dataset[0]
print(f"text   : {s['text'].shape}")    # → (300,)
print(f"audio  : {s['audio'].shape}")   # → (74,)
print(f"vision : {s['vision'].shape}")  # → (35,)
print(f"valence: {s['valence'].item():.4f}")
print(f"arousal: {s['arousal'].item():.4f}")
````

````python
from torch.utils.data import random_split, DataLoader

# ── Split ─────────────────────────────────────────────────
total = len(dataset)
n_train = int(0.8 * total)   # 8,730
n_val   = int(0.1 * total)   #  1,091
n_test  = total - n_train - n_val  # 1,092

train_set, val_set, test_set = random_split(
    dataset, [n_train, n_val, n_test],
    generator=torch.Generator().manual_seed(42)
)

# ── Collate fn (handles dict samples) ─────────────────────
def collate_fn(batch):
    return {
        'text':    torch.stack([s['text']    for s in batch]),   # (B, 300)
        'audio':   torch.stack([s['audio']   for s in batch]),   # (B, 74)
        'vision':  torch.stack([s['vision']  for s in batch]),   # (B, 35)
        'valence': torch.stack([s['valence'] for s in batch]),   # (B,)
        'arousal': torch.stack([s['arousal'] for s in batch]),   # (B,)
    }

train_loader = DataLoader(train_set, batch_size=64, shuffle=True,  collate_fn=collate_fn)
val_loader   = DataLoader(val_set,   batch_size=64, shuffle=False, collate_fn=collate_fn)
test_loader  = DataLoader(test_set,  batch_size=64, shuffle=False, collate_fn=collate_fn)

# ── Verify ────────────────────────────────────────────────
batch = next(iter(train_loader))
for k, v in batch.items():
    print(f"{k:8s}: {v.shape}")
````

````python
import torch.nn as nn

class ModalEncoder(nn.Module):
    """Projects each modality to a shared hidden dim."""
    def __init__(self, input_dim, hidden_dim=128, dropout=0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, hidden_dim),
        )
    def forward(self, x):
        return self.net(x)   # (B, hidden_dim)


class MultimodalFusionModel(nn.Module):
    def __init__(self, hidden_dim=128, n_heads=4, dropout=0.3):
        super().__init__()

        # Per-modality encoders
        self.text_enc   = ModalEncoder(300, hidden_dim, dropout)
        self.audio_enc  = ModalEncoder(74,  hidden_dim, dropout)
        self.vision_enc = ModalEncoder(35,  hidden_dim, dropout)

        # Transformer fusion over 3 tokens (text, audio, vision)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim, nhead=n_heads,
            dim_feedforward=256, dropout=dropout,
            batch_first=True
        )
        self.fusion = nn.TransformerEncoder(encoder_layer, num_layers=2)

        # VA regression head
        self.va_head = nn.Sequential(
            nn.Linear(hidden_dim * 3, 64),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(64, 2)    # → [valence, arousal]
        )

    def forward(self, text, audio, vision):
        t = self.text_enc(text).unsqueeze(1)    # (B, 1, H)
        a = self.audio_enc(audio).unsqueeze(1)  # (B, 1, H)
        v = self.vision_enc(vision).unsqueeze(1)# (B, 1, H)

        tokens = torch.cat([t, a, v], dim=1)    # (B, 3, H)
        fused  = self.fusion(tokens)             # (B, 3, H)

        flat = fused.reshape(fused.size(0), -1) # (B, 3*H)
        out  = self.va_head(flat)               # (B, 2)

        valence = out[:, 0]
        arousal = out[:, 1]
        return valence, arousal


# ── Sanity check ──────────────────────────────────────────
model = MultimodalFusionModel(hidden_dim=128)
batch = next(iter(train_loader))

v_pred, a_pred = model(batch['text'], batch['audio'], batch['vision'])
print(f"valence_pred: {v_pred.shape}")   # → (64,)
print(f"arousal_pred: {a_pred.shape}")   # → (64,)
print(f"Params: {sum(p.numel() for p in model.parameters()):,}")
````

````python
import torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingLR

# ── CCC metric ────────────────────────────────────────────
def concordance_correlation_coefficient(pred, target):
    """CCC = 2σ_xy / (σ_x² + σ_y² + (μ_x - μ_y)²)"""
    pred_mean   = pred.mean()
    target_mean = target.mean()
    pred_var    = pred.var()
    target_var  = target.var()
    covariance  = ((pred - pred_mean) * (target - target_mean)).mean()
    ccc = (2 * covariance) / (pred_var + target_var + (pred_mean - target_mean) ** 2 + 1e-8)
    return ccc

# ── Loss: MSE + CCC jointly ───────────────────────────────
def va_loss(v_pred, a_pred, v_true, a_true, alpha=0.5):
    mse_v = nn.functional.mse_loss(v_pred, v_true)
    mse_a = nn.functional.mse_loss(a_pred, a_true)
    ccc_v = 1 - concordance_correlation_coefficient(v_pred, v_true)
    ccc_a = 1 - concordance_correlation_coefficient(a_pred, a_true)
    return alpha * (mse_v + mse_a) + (1 - alpha) * (ccc_v + ccc_a)

# ── Setup ─────────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

model     = MultimodalFusionModel(hidden_dim=128).to(device)
optimizer = optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=30, eta_min=1e-6)

# ── Training loop ─────────────────────────────────────────
def run_epoch(loader, train=True):
    model.train() if train else model.eval()
    total_loss, ccc_v_sum, ccc_a_sum = 0, 0, 0

    ctx = torch.enable_grad() if train else torch.no_grad()
    with ctx:
        for batch in loader:
            text    = batch['text'].to(device)
            audio   = batch['audio'].to(device)
            vision  = batch['vision'].to(device)
            v_true  = batch['valence'].to(device)
            a_true  = batch['arousal'].to(device)

            v_pred, a_pred = model(text, audio, vision)
            loss = va_loss(v_pred, a_pred, v_true, a_true)

            if train:
                optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()

            total_loss += loss.item()
            ccc_v_sum  += concordance_correlation_coefficient(v_pred.detach(), v_true).item()
            ccc_a_sum  += concordance_correlation_coefficient(a_pred.detach(), a_true).item()

    n = len(loader)
    return total_loss / n, ccc_v_sum / n, ccc_a_sum / n

# ── Train ─────────────────────────────────────────────────
EPOCHS = 30
best_val_loss = float('inf')
history = []

for epoch in range(1, EPOCHS + 1):
    tr_loss, tr_ccc_v, tr_ccc_a = run_epoch(train_loader, train=True)
    vl_loss, vl_ccc_v, vl_ccc_a = run_epoch(val_loader,   train=False)
    scheduler.step()

    history.append({
        'epoch': epoch,
        'tr_loss': tr_loss, 'vl_loss': vl_loss,
        'tr_ccc_v': tr_ccc_v, 'vl_ccc_v': vl_ccc_v,
        'tr_ccc_a': tr_ccc_a, 'vl_ccc_a': vl_ccc_a,
    })

    # Save best checkpoint
    if vl_loss < best_val_loss:
        best_val_loss = vl_loss
        torch.save(model.state_dict(), "best_model.pt")
        tag = " ← best"
    else:
        tag = ""

    if epoch % 5 == 0 or epoch == 1:
        print(
            f"Ep {epoch:02d} | "
            f"Loss {tr_loss:.4f}/{vl_loss:.4f} | "
            f"CCC-V {tr_ccc_v:.3f}/{vl_ccc_v:.3f} | "
            f"CCC-A {tr_ccc_a:.3f}/{vl_ccc_a:.3f}{tag}"
        )

print(f"\nBest val loss: {best_val_loss:.4f}")
````

````python
def pool_features(features, method='mean'):
    # Replace Inf/-Inf with NaN, then nanmean/nanmax ignores them
    features = features.copy()
    features[~np.isfinite(features)] = np.nan

    if method == 'mean':
        result = np.nanmean(features, axis=0)
    elif method == 'max':
        result = np.nanmax(features, axis=0)

    # If an entire feature column was Inf/NaN, nanmean returns NaN → replace with 0
    result = np.nan_to_num(result, nan=0.0, posinf=0.0, neginf=0.0)
    return result


class MOSEIDataset(Dataset):
    def __init__(self, mosei_dict, pool='mean'):
        self.samples = []

        for video_id, utterances in mosei_dict.items():
            for seg in utterances:
                try:
                    labels_raw  = np.array(seg['Labels']).item()
                    covarep_raw = np.array(seg['COVAREP']).item()
                    facet_raw   = np.array(seg['FACET']).item()
                    wordvec_raw = np.array(seg['WORDVEC']).item()

                    text   = pool_features(wordvec_raw['features'].astype(np.float32), pool)
                    audio  = pool_features(covarep_raw['features'].astype(np.float32), pool)
                    vision = pool_features(facet_raw['features'].astype(np.float32),   pool)

                    label_feats = labels_raw['features']
                    valence, arousal = extract_va(label_feats)

                    # Skip if labels themselves are non-finite
                    if not (np.isfinite(valence) and np.isfinite(arousal)):
                        continue

                    self.samples.append({
                        'text':    torch.tensor(text),
                        'audio':   torch.tensor(audio),
                        'vision':  torch.tensor(vision),
                        'valence': torch.tensor(valence, dtype=torch.float32),
                        'arousal': torch.tensor(arousal, dtype=torch.float32),
                        'video_id': video_id,
                    })

                except Exception as e:
                    print(f"Skipped {video_id}: {e}")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        return self.samples[idx]


# ── Rebuild everything from scratch ───────────────────────
dataset = MOSEIDataset(mosei)
print(f"Total utterances after cleaning: {len(dataset)}")

# Verify no Infs remain
bad = sum(
    1 for i in range(len(dataset))
    for key in ['text', 'audio', 'vision']
    if not torch.isfinite(dataset[i][key]).all()
)
print(f"Remaining non-finite samples: {bad}")

# Rebuild splits and loaders
total   = len(dataset)
n_train = int(0.8 * total)
n_val   = int(0.1 * total)
n_test  = total - n_train - n_val

train_set, val_set, test_set = random_split(
    dataset, [n_train, n_val, n_test],
    generator=torch.Generator().manual_seed(42)
)

train_loader = DataLoader(train_set, batch_size=64, shuffle=True,  collate_fn=collate_fn)
val_loader   = DataLoader(val_set,   batch_size=64, shuffle=False, collate_fn=collate_fn)
test_loader  = DataLoader(test_set,  batch_size=64, shuffle=False, collate_fn=collate_fn)

# Quick forward check
model = MultimodalFusionModel(hidden_dim=128).to(device)
batch = next(iter(train_loader))
with torch.no_grad():
    v_pred, a_pred = model(batch['text'].to(device), batch['audio'].to(device), batch['vision'].to(device))
    print(f"Pred NaNs: {torch.isnan(v_pred).any()} | Pred range V: {v_pred.min():.3f} to {v_pred.max():.3f}")
````

````python
import torch
import torch.nn as nn

# ── Temporal encoder: replaces pool_features for audio/vision ──
class TemporalEncoder(nn.Module):
    """
    Takes (B, T, D) sequence → (B, H) summary via GRU + attention pooling.
    Much more expressive than mean-pooling.
    """
    def __init__(self, input_dim, hidden_dim=128, num_layers=2, dropout=0.3):
        super().__init__()
        self.gru = nn.GRU(
            input_dim, hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.attn = nn.Linear(hidden_dim * 2, 1)  # *2 for bidirectional
        self.proj = nn.Linear(hidden_dim * 2, hidden_dim)
        self.norm = nn.LayerNorm(hidden_dim)

    def forward(self, x, lengths=None):
        # x: (B, T, D)
        out, _ = self.gru(x)                     # (B, T, H*2)
        weights = torch.softmax(self.attn(out), dim=1)  # (B, T, 1)
        pooled  = (out * weights).sum(dim=1)     # (B, H*2)
        return self.norm(self.proj(pooled))       # (B, H)


# ── Updated Dataset — keep sequences, pad to fixed length ──
def pad_sequence_2d(features, max_len):
    """Pad or truncate (T, D) → (max_len, D)."""
    T, D = features.shape
    if T >= max_len:
        return features[:max_len]
    pad = np.zeros((max_len - T, D), dtype=np.float32)
    return np.vstack([features, pad])

def clean_features(features):
    """Replace non-finite values with 0."""
    features = features.copy()
    features[~np.isfinite(features)] = 0.0
    return features

# Max lengths (use 95th percentile to avoid extreme padding)
MAX_TEXT   = 40    # WORDVEC  frames
MAX_AUDIO  = 600   # COVAREP  frames
MAX_VISION = 200   # FACET    frames

class MOSEITemporalDataset(torch.utils.data.Dataset):
    def __init__(self, mosei_dict):
        self.samples = []

        for video_id, utterances in mosei_dict.items():
            for seg in utterances:
                try:
                    labels_raw  = np.array(seg['Labels']).item()
                    covarep_raw = np.array(seg['COVAREP']).item()
                    facet_raw   = np.array(seg['FACET']).item()
                    wordvec_raw = np.array(seg['WORDVEC']).item()

                    text   = clean_features(wordvec_raw['features'].astype(np.float32))
                    audio  = clean_features(covarep_raw['features'].astype(np.float32))
                    vision = clean_features(facet_raw['features'].astype(np.float32))

                    # Pad sequences
                    text   = pad_sequence_2d(text,   MAX_TEXT)    # (40,  300)
                    audio  = pad_sequence_2d(audio,  MAX_AUDIO)   # (600, 74)
                    vision = pad_sequence_2d(vision, MAX_VISION)  # (200, 35)

                    label_feats = labels_raw['features']
                    valence, arousal = extract_va(label_feats)
                    if not (np.isfinite(valence) and np.isfinite(arousal)):
                        continue

                    self.samples.append({
                        'text':    torch.tensor(text),
                        'audio':   torch.tensor(audio),
                        'vision':  torch.tensor(vision),
                        'valence': torch.tensor(valence, dtype=torch.float32),
                        'arousal': torch.tensor(arousal, dtype=torch.float32),
                    })

                except Exception as e:
                    print(f"Skipped {video_id}: {e}")

    def __len__(self): return len(self.samples)
    def __getitem__(self, idx): return self.samples[idx]


# ── Updated model ──────────────────────────────────────────
class MultimodalTemporalModel(nn.Module):
    def __init__(self, hidden_dim=128, n_heads=4, dropout=0.3):
        super().__init__()

        # Sequence encoders (replace flat ModalEncoder)
        self.text_enc   = TemporalEncoder(300, hidden_dim, dropout=dropout)
        self.audio_enc  = TemporalEncoder(74,  hidden_dim, dropout=dropout)
        self.vision_enc = TemporalEncoder(35,  hidden_dim, dropout=dropout)

        # Transformer fusion (same as before)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim, nhead=n_heads,
            dim_feedforward=256, dropout=dropout,
            batch_first=True
        )
        self.fusion = nn.TransformerEncoder(encoder_layer, num_layers=2)

        # VA head
        self.va_head = nn.Sequential(
            nn.Linear(hidden_dim * 3, 64),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(64, 2)
        )

    def forward(self, text, audio, vision):
        t = self.text_enc(text).unsqueeze(1)     # (B, 1, H)
        a = self.audio_enc(audio).unsqueeze(1)   # (B, 1, H)
        v = self.vision_enc(vision).unsqueeze(1) # (B, 1, H)

        tokens = torch.cat([t, a, v], dim=1)     # (B, 3, H)
        fused  = self.fusion(tokens)              # (B, 3, H)
        flat   = fused.reshape(fused.size(0), -1)# (B, 3*H)
        out    = self.va_head(flat)               # (B, 2)

        return out[:, 0], out[:, 1]              # valence, arousal


# ── Build and verify ───────────────────────────────────────
temporal_dataset = MOSEITemporalDataset(mosei)
print(f"Dataset size: {len(temporal_dataset)}")

s = temporal_dataset[0]
print(f"text   : {s['text'].shape}")    # (40,  300)
print(f"audio  : {s['audio'].shape}")   # (600, 74)
print(f"vision : {s['vision'].shape}")  # (200, 35)

total   = len(temporal_dataset)
n_train = int(0.8 * total)
n_val   = int(0.1 * total)
n_test  = total - n_train - n_val

train_t, val_t, test_t = torch.utils.data.random_split(
    temporal_dataset, [n_train, n_val, n_test],
    generator=torch.Generator().manual_seed(42)
)

train_loader_t = torch.utils.data.DataLoader(train_t, batch_size=32, shuffle=True,  collate_fn=collate_fn)
val_loader_t   = torch.utils.data.DataLoader(val_t,   batch_size=32, shuffle=False, collate_fn=collate_fn)
test_loader_t  = torch.utils.data.DataLoader(test_t,  batch_size=32, shuffle=False, collate_fn=collate_fn)

# Sanity check forward pass
model_t = MultimodalTemporalModel(hidden_dim=128).to(device)
batch   = next(iter(train_loader_t))
with torch.no_grad():
    vp, ap = model_t(batch['text'].to(device), batch['audio'].to(device), batch['vision'].to(device))

print(f"\nForward pass OK — v:{vp.shape} a:{ap.shape}")
print(f"Params: {sum(p.numel() for p in model_t.parameters()):,}")
````
